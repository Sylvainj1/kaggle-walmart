```python
import pandas as pd
```

```python
ca1= pd.read_csv('../../data/prophet_store/prophet_CA1.csv')
ca2= pd.read_csv('../../data/prophet_store/prophet_CA2.csv')
ca3= pd.read_csv('../../data/prophet_store/prophet_CA3.csv')
ca4= pd.read_csv('../../data/prophet_store/prophet_CA4.csv')

tx1= pd.read_csv('../../data/prophet_store/prophet_TX1.csv')
tx2= pd.read_csv('../../data/prophet_store/prophet_TX2.csv')
tx3= pd.read_csv('../../data/prophet_store/prophet_TX3.csv')

wi1= pd.read_csv('../../data/prophet_store/prophet_WI1.csv')
wi2= pd.read_csv('../../data/prophet_store/prophet_WI2.csv')
wi3= pd.read_csv('../../data/prophet_store/prophet_WI3.csv')

ca1.drop('Unnamed: 0', axis=1, inplace = True)
ca2.drop('Unnamed: 0', axis=1, inplace = True)
ca3.drop('Unnamed: 0', axis=1, inplace = True)
ca4.drop('Unnamed: 0', axis=1, inplace = True)
tx1.drop('Unnamed: 0', axis=1, inplace = True)
```

```python
df_prophet_all = pd.concat([ca1,ca2,ca3,ca4,tx1,tx2,tx3,wi1,wi2,wi3],ignore_index=True)
```

```python
df_prophet_all
```

```python
df_validation= df_prophet_all.iloc[:,0:29]
df_validation
```

```python
df_evaluation= df_prophet_all.iloc[:,29:]
df_evaluation.columns= df_validation.columns[1:]

df_evaluation.insert(0, 'id', df_validation["id"].str.replace("validation$", "evaluation"))
df_evaluation
```

```python
submission_prophet_store = pd.concat([df_validation,df_evaluation], ignore_index= True)
```

```python
submission_prophet_store
```

```python
submission_prophet_store.set_index('id', inplace= True)
```
